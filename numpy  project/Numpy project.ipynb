#### Convert to numpy array

```python
units = np.array(df["Units_Sold"])
price = np.array(df["Avg_Price_EUR"])
revenue = np.array(df["Revenue_EUR"])
gdp = np.array(df["GDP_Growth"])
fuel = np.array(df["Fuel_Price_Index"])
```

### Data Exploration

```python
print("Total Records:", units.size)
print("Average Units Sold:", np.mean(units))
print("Maximum Units Sold:", np.max(units))
print("Minimum Units Sold:", np.min(units))
```

### Revenue Analysis

```python
print("Total Revenue:", np.sum(revenue))
print("Average Revenue:", np.mean(revenue))
print("Revenue Standard Deviation:", np.std(revenue))
```

### Price Analysis

```python
print("Average Car Price:", np.mean(price))
print("Highest Price:", np.max(price))
print("Lowest Price:", np.min(price))
```

### sorting


```python
sorted_units = np.sort(units)
print("Sorted Units Sold:", sorted_units)
```

```python
top_sales = np.sort(units)[-5:]
print("Top 5 Sales:", top_sales)
```
