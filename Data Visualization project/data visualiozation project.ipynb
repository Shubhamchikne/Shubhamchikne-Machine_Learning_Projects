### Theory of Data Visualization
Introduction

Data Visualization is the graphical representation of data using charts, graphs, and plots. It helps to understand patterns, trends, relationships, and insights from data in a visual format.

Data visualization makes complex data easy to understand and helps in better decision-making in business, research, and data analysis.

Data visualization is an important part of Data Science and Data Analytics.

Objectives of Data Visualization

The main objectives of data visualization are:

To represent data visually
To identify trends and patterns
To compare data values
To analyze relationships between variables
To detect outliers
To support decision making
Types of Data Visualization
1. Line Chart

A line chart is used to show trends over time or ordered data. It connects data points using lines.

Example: CGPA vs Salary, Sales over Years

2. Bar Chart

A bar chart is used to compare different categories of data.

Example: Number of placed vs unplaced students

3. Histogram

A histogram is used to show the frequency distribution of numerical data. It shows how data is distributed into different ranges (bins).

4. Scatter Plot

A scatter plot is used to show the relationship between two numerical variables. It helps to identify correlation between variables.

Example: CGPA vs Salary, Skills vs Salary

Related concept: Correlation

5. Box Plot

A box plot is used to show data distribution and detect outliers. It shows the five-number summary:

Minimum
First Quartile (Q1)
Median
Third Quartile (Q3)
Maximum

Related concept: Outlier

Advantages of Data Visualization
Easy to understand data
Saves time
Helps in decision making
Identifies trends and patterns
Detects outliers and errors
Useful in business and research
Applications of Data Visualization

Data visualization is used in:

Business analytics
Finance

Education

Machine Learning
Conclusion

Data visualization is an important technique used to represent data in graphical form. It helps to understand trends, patterns, and relationships between variables. ### Line plot

```python
plt.plot(df["cgpa"])
plt.xlabel("Cgpa")
plt.ylabel("salary_package_lpa")
plt.title("CGPA")
plt.show()
```

```python
plt.plot(df["projects_count"], df["salary_package_lpa"])
plt.xlabel("Projects Count")
plt.ylabel("Salary")
plt.title("Projects vs Salary")
plt.show()
```

### Histogram 

```python
plt.hist(df["salary_package_lpa"])
plt.title("Salary Package Distribution")
plt.xlabel("Salary (LPA)")
plt.ylabel("Number of Students")
plt.show()
```

```python
plt.scatter(df["cgpa"], df["salary_package_lpa"])
plt.title("CGPA vs Salary")
plt.xlabel("CGPA")
plt.ylabel("Salary (LPA)")
plt.show()
```

```python
plt.scatter(df["internships"], df["salary_package_lpa"])
plt.title("Internships vs Salary")
plt.show()
```

```python
plt.boxplot(df["salary_package_lpa"])
plt.title("Box Plot of Salary Package")
plt.show()
```
