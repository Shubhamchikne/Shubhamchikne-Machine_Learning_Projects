### THEORY: Exploratory Data Analysis (EDA)
Introduction

Exploratory Data Analysis (EDA) is an approach used in data analysis to summarize, visualize, and understand the main characteristics of a dataset. It helps in discovering patterns, detecting anomalies, testing assumptions, and checking relationships between variables before applying machine learning models or statistical techniques.

EDA is an important step in the data analysis process because real-world data is often incomplete, inconsistent, and noisy.

Objectives of EDA

The main objectives of Exploratory Data Analysis are:

To understand the structure of the dataset
To identify missing values and duplicate data
To detect outliers and errors
To understand relationships between variables
To summarize the dataset using statistics
To prepare data for further analysis or modeling
Steps in EDA Process
1. Data Collection

Data is collected from various sources such as CSV files, Excel files, databases, or websites.

2. Data Cleaning

In this step, we:

Handle missing values
Remove duplicate records
Fix incorrect data types
Handle outliers

Data cleaning improves data quality.

3. Data Understanding

This includes:

Checking number of rows and columns
Understanding data types
Viewing top and bottom records
Summary statistics

Common functions used:

head()
tail()
info()
describe()
4. Data Visualization

Visualization helps to understand data using graphs such as:

Bar chart
Histogram
Pie chart
Box plot
Scatter plot
Heatmap

Visualization makes it easier to identify trends and relationships.

5. Finding Relationships

EDA helps in finding relationships between variables using:

Correlation
Pair plots
Grouped analysis
Tools Used in EDA

Common Python libraries used:

Pandas – for data manipulation
NumPy – for numerical operations
Matplotlib – for plotting graphs
Seaborn – for advanced visualization
Importance of EDA
Helps in better understanding of data
Improves model accuracy
Helps in feature selection
Detects errors before modeling
Saves time in later stages
Conclusion

Exploratory Data Analysis is a crucial step in any data science project. It helps to understand the dataset, clean the data, and extract useful insights using statistical methods and visualization techniques. Proper EDA leads to better decision-making and improved model performance.

### Convert Postal Code to string

```python
df['Postal Code'] = df['Postal Code'].astype(str)
```

```python
df["Postal Code"].dtype
```

```python
df.isnull().sum()
```

```python
df['Postal Code'] = df['Postal Code'].fillna( 'Unknown')
```

```python
df.isnull().sum()
```

### Convert dates

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d/%m/%Y')
```

### Create Delivery Time feature

```python
df['Delivery Time'] = (df['Ship Date'] - df['Order Date']).dt.days
```

```python
region_sales = df.groupby('Region')['Sales'].sum()
region_sales.plot(kind='bar', title='Sales by Region')
plt.show()
```

```python
df['Sales'].describe()
```

```python
top_customers = df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(10)
top_customers.plot(kind='barh', title='Top 10 Customers')
plt.show()
```

```python
sns.boxplot(x='Ship Mode', y='Sales', data=df)
plt.title("Sales Distribution by Ship Mode")
plt.show()
```

```python
numeric_cols = ['Sales', 'Delivery Time']
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm')
plt.show()
```
